```python
from apply_filter import applyFilter as fil
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("sneaker.png")
```

```python
plt.imshow(img)
plt.title('Regular Sneaker')
plt.show()
```

```python
edgeFilter = np.array([[-1, -1, -1],
                          [-1, 8, -1],
                          [-1, -1, -1]])
```

```python
filteredImg = fil(img, edgeFilter)
```

```python
plt.imshow(filteredImg)
plt.title('Manually Applying Filter')
plt.show()
```

```python
copyOfImg = img.copy()
gray = cv2.cvtColor(copyOfImg, cv2.COLOR_BGR2GRAY)
dst = cv2.filter2D(gray, -1, edgeFilter)
```

```python
plt.imshow(dst)
plt.title('Upper Left Corner of Img')
plt.show()
```

```python
basicFilter = np.array([[2, 2, 2],
                       [2, 2, 2],
                       [2, 2, 2]])
```

```python
basicImg = fil(img, basicFilter)
```

```python
plt.imshow(basicImg)
plt.title('Upper Left Corner of Img')
plt.show()
```
