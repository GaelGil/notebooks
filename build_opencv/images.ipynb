```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# read in image
img = cv2.imread("sneaker.png")
```

```python
plt.imshow(img)
plt.title('Regular Sneaker')
plt.show()
```

```python
# gray gilter img
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# blury img
img_blur = cv2.GaussianBlur(img, (7,7), 0)

# cool filter img 
# TODO: understand how canny works
img_canny = cv2.Canny(img, 100, 100)
```

```python
plt.imshow(gray)
plt.title('Regular Sneaker')
plt.show()
```

```python
plt.imshow(img_blur)
plt.title('Blur Sneaker')
plt.show()
```

```python
plt.imshow(img_canny)
plt.title('Filter')
plt.show()
```

```python
b = img.copy()
# set green and red channels to 0
b[:, :, 1] = 0
b[:, :, 2] = 0


g = img.copy()
# set blue and red channels to 0
g[:, :, 0] = 0
g[:, :, 2] = 0

r = img.copy()
# set blue and green channels to 0
r[:, :, 0] = 0
r[:, :, 1] = 0
```

```python
plt.imshow(b)
plt.title('Red Sneaker')
plt.show()
```

```python
plt.imshow(g)
plt.title('Green Sneaker')
plt.show()
```

```python
plt.imshow(r)
plt.title('Blue Sneaker')
plt.show()
```

```python
custom_filter = np.array([[1, 0, 0],
                          [0, 1, 0],
                          [0, 0, 1]])
```

```python
dst = cv2.filter2D(img, -100, custom_filter)
plt.imshow(dst)
plt.title('Upper Left Corner of Img')
plt.show()
```

```python
other_filter = np.array([[0, -1, 0],
                          [-1, 4, -1],
                          [0, -1, 0]])
```
